```python
import pandas as pd
import numpy as np
import gspread
import datetime
import time
import schedule
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from datetime import timedelta
  
def weather():
    SHEET_ID = '1LkG3mwIDkBbsnpm4xNGJqvXSOiOeGGqnqlreI0BDmIs'
    SHEET_NAME = 'Sheet1'
    gc = gspread.service_account('C:/Users/KIIT/Documents/6th Sem/Project/weather-382617-007311942bed.json')
    spreadsheet = gc.open_by_key(SHEET_ID)
    worksheet = spreadsheet.worksheet(SHEET_NAME)
    rows = worksheet.get_all_records()
    df = pd.DataFrame(rows)
    
    df["Date"]=pd.to_datetime(df["Date"])
    
    df["target_temp"] = df.shift(-1)["Temperature"]
    df["target_press"] = df.shift(-1)["Pressure"]
    df["target_humid"] = df.shift(-1)["Humidity"]
    
    obs=df.iloc[-1].drop(['Date','target_temp','target_press','target_humid'])
    dropped_df=df.dropna()
    
    X_temp=dropped_df.drop(['target_temp','target_press','target_humid','Date'], axis=1)
    y_temp=dropped_df['target_temp']
    X_press=dropped_df.drop(['target_temp','target_press','target_humid','Date'], axis=1)
    y_press=dropped_df['target_press']
    X_humid=dropped_df.drop(['target_temp','target_press','target_humid','Date'], axis=1)
    y_humid=dropped_df['target_humid']
    
    X_temp_train, X_temp_test, y_temp_train, y_temp_test=train_test_split(X_temp,y_temp,test_size=0.3, random_state=123)
    X_press_train, X_press_test, y_press_train, y_press_test=train_test_split(X_press,y_press,test_size=0.3, random_state=123)
    X_humid_train, X_humid_test, y_humid_train, y_humid_test=train_test_split(X_humid,y_humid,test_size=0.3, random_state=123)
    
    reg_temp = Ridge(alpha=.1)
    reg_press = Ridge(alpha=.1)
    reg_humid = Ridge(alpha=.1)
    
    reg_temp.fit(X_temp_train,y_temp_train)
    reg_press.fit(X_press_train,y_press_train)
    reg_humid.fit(X_humid_train,y_humid_train)
    
    print("Current Weather :\n"+str(df["Date"].iloc[-1])+"\nTemperature: "
          +str(float(obs["Temperature"]))+"\nPressure: "+str(float(obs["Pressure"]))
          +"\nHumidity: "+str(float(obs["Humidity"]))+"\n\nPredictions:")
    
    for i in range(13):
        if(i!=0):
            timey=df['Date'].iloc[-1]+timedelta(minutes=5*i)
            print("\n"+str(timey)+" :")
            temp=float(reg_temp.predict([obs]))
            press=float(reg_press.predict([obs]))
            humid=float(reg_humid.predict([obs]))
            print("Temperature: "+str(temp)+"\nPressure: "+str(press)+"\nHumidity: "+str(humid)+"\n")
            obs["Temperature"]=temp
            obs["Pressure"]=press
            obs["Humidity"]=humid
            
schedule.every(5).minutes.do(weather)
  
while True:
    schedule.run_pending()
    time.sleep(1)
```
